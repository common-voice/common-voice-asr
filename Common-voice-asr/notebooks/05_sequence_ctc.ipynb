```python
# Decoding a batch and printing sample transcripts + WER
import torch
import torchaudio
import torch.nn.functional as F
from neural_networks.datasets import CTC_MiniCVDataset, ctc_collate_fn, char_to_id
from dotenv import load_dotenv
from torch.utils.data import DataLoader
from neural_networks.cnn_encoder import CTC_CNNEncoder
from neural_networks.wrap_encoder import WrapEncoder
from neural_networks.greedy_ctc_decoder import GreedyCTCDecoder
from jiwer import wer

load_dotenv()
BASE_DIR = Path(os.getenv("BASE_DIR"))
manifest_path = os.path.join(BASE_DIR, "data/manifest_full.csv")
spect_dir = os.path.join(BASE_DIR, "data/processed/full_mini_cv")

dataset = CTC_MiniCVDataset(manifest_path, spect_dir)
loader = DataLoader(dataset, batch_size=1, collate_fn=ctc_collate_fn, shuffle=True)

encoder = CTC_CNNEncoder()
model = WrapEncoder(encoder, 36, False)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()

char_id_dict = char_to_id()
id_to_char_dict = {idx: char for char, idx in char_id_dict.items()}
id_to_char_list = [char for idx, char in sorted(id_to_char_dict.items())]
decoder = GreedyCTCDecoder(id_to_char_list)

batch = next(iter(loader))
inputs, targets, input_lengths, target_lengths = batch
inputs = inputs.to(device)
with torch.no_grad():
    logits = model(inputs)
    log_probs = F.log_softmax(logits, dim=2)
    hypotheses = decoder(log_probs.cpu())

references = []
for i, length in enumerate(target_lengths):
    target_seq = targets[sum(target_lengths[:i]):sum(target_lengths[:i+1])]
    chars = [id_to_char_dict[idx.item()] for idx in target_seq]
    references.append("".join(chars))

for ref, hyp in zip(references, hypotheses):
    ref_words = ref.strip().split()
    hyp_words = hyp
    print(f"Reference:  {' '.join(ref_words)}")
    print(f"Hypothesis: {' '.join(hyp_words)}")
    wer = torchaudio.functional.edit_distance(ref_words, hyp_words) / max(len(ref_words), 1)
    print(f"WER: {wer:.2f}\n")
```

```python
# Plotting loss vs. epoch and WER vs. epoch from TensorBoard logs.
import matplotlib.pyplot as plt
import os
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator


def plot_loss_curves(log_dir, label_pref=""):
    event_acc = EventAccumulator(log_dir)
    event_acc.Reload()

    train_losses = event_acc.Scalars("train/ctc_loss")
    val_losses = event_acc.Scalars("val/ctc_loss")

    train_steps = [x.step for x in train_losses]
    train_vals = [x.value for x in train_losses]
    val_steps = [x.step for x in val_losses]
    val_vals = [x.value for x in val_losses]

    plt.figure(figsize=(10,5))
    plt.plot(train_steps, train_vals, label=f"{label_pref}Train loss")
    plt.plot(val_steps, val_vals, label=f"{label_pref}Val loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training vs Validation CTCLoss")
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_wer_curves(log_dir, label_pref=""):
    event_acc = EventAccumulator(log_dir)
    event_acc.Reload()

    train_wer = event_acc.Scalars("train/wer")
    val_wer = event_acc.Scalars("val/wer")

    train_steps = [x.step for x in train_wer]
    train_vals = [x.value for x in train_wer]
    val_steps = [x.step for x in val_wer]
    val_vals = [x.value for x in val_wer]

    plt.figure(figsize=(10,5))
    plt.plot(train_steps, train_vals, label=f"{label_pref}Train WER")
    plt.plot(val_steps, val_vals, label=f"{label_pref}Val WER")
    plt.xlabel("Epoch")
    plt.ylabel("WER")
    plt.title("Training vs Validation WER")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_loss_curves("/Users/setongerrity/Desktop/Mozilla/common-voice-asr/Common-voice-asr/neural_networks/runs/week4_ctc", "CTC")
plot_loss_curves("/Users/setongerrity/Desktop/Mozilla/common-voice-asr/Common-voice-asr/neural_networks/runs/week4_ctc", "CTC")
```
